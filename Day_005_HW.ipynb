```python
app_training=pd.DataFrame(app_train)
app_training.head(50)
```

```python
print('AMT_INCOME_TOTAL的平均值是: ',app_training['AMT_INCOME_TOTAL'].mean()) #在 AMT_INCOME_TOTAL 所有數值的平均值
print('AMT_CREDIT的平均值是: ',app_training['AMT_CREDIT'].mean())
```

```python
print('AMT_INCOME_TOTAL的標準差是: ',app_training['AMT_INCOME_TOTAL'].std())
print('AMT_ANNUITY的標準差是: ',app_training['AMT_ANNUITY'].std())
```

```python
app_training['AMT_ANNUITY'].plot.hist(bins=100,facecolor='red',edgecolor='black', alpha=0.8,title='AMT_ANNUITY')
```

```python
app_training['AMT_CREDIT'].plot.hist(bins=100,facecolor='blue',edgecolor='black', alpha=0.6,title='AMT_CREDIT')
```
